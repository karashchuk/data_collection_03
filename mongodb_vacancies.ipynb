#### 1) Развернуть у себя на компьютере/виртуальной машине/хостинге MongoDB и реализовать функцию, записывающую собранные вакансии в созданную БД

```python
#!pip install pymongo
from pymongo import MongoClient
from pprint import pprint
import json
import numpy as np
import pandas as pd
```

```python
#подключимся к базе и создадим две коллекции
client = MongoClient('localhost', 27017)
db = client['vacancies']
mc1 = db.mycollection1
mc2 = db.mycollection2
```

```python
#считаем данные с с файла в формате .csv
df = pd.read_csv('mytest.csv', sep=';')
df.head()
```

```python
mydata = df.to_dict(orient='records')
pprint(mydata[:2])
```

```python
#загрузим в базу в первую коллекцию
mc1.insert_many(mydata)
```

```python
#проверим что в базе:
mylist =list(mc1.find({}))
mylist[:3]
```

```python
# Можно аналогично загрузить из json
with open("data_file.json", "r") as f:
    data = json.loads(f.read())
pprint(data[:2])
mc2.insert_many(data)
list(mc2.find({}))[:2]
```

#### 2) Написать функцию, которая производит поиск и выводит на экран вакансии с заработной платой больше введенной суммы

```python
sum_min = int(input('Введите зарплату, минимально от которой будут вакансии'))
for vacancy in mc1.find({'comp_min': {'$gte': sum_min }}):
    pprint(vacancy)
```

```python
#сделаем вывод в виде DataFrame
sum_min = int(input('Введите зарплату, минимально от которой будут вакансии'))
pd.DataFrame(list(mc1.find({'comp_min': {'$gte': sum_min }},{'_id':0})))
```

#### 3*)Написать функцию, которая будет добавлять в вашу базу данных только новые вакансии с сайта

```python
# Сделаем тестовую коллекцию для наглядности обработки обновления вакансий и загрузим туда 5 вакансий
mytest = db.mytest
mytest.insert_many((mydata[:5]))
pd.DataFrame(list(mytest.find({},{'_id': 0})))
```

```python
#создадим список из вакансий в которых есть и указанные выше, 
#а также в существующих изменим некоторые поля
newdata = mydata[:8]
newdata[0]['comp_min'] = 150000
newdata[2]['comp_max'] = 211000
newdata[3]['address'] = 'Тула'
pd.DataFrame(newdata)
```

```python
#тогда новые данные будут добавляться если не нашлись в базе или обновляться если нашлись:
for v in newdata:
    mytest.update_one({'href': v['href']},{'$set': v},upsert=True)
pd.DataFrame(list(mytest.find({},{'_id': 0})))

```
